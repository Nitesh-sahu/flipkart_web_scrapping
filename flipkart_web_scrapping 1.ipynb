```python
#  import useful libraries
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as uReq
```

```python
# website url
flip_url="https://www.flipkart.com/search?q="
```

```python
# Take a input for the user and 
# if there is space in between the serch input the rplace it by + 
# then concatenet the website_url and inputed url
# then store the usl in a variable for refernce

search_product=input("Enter product to search")
search_product=search_product.replace(" ","+")
fin_url=flip_url+search_product
print(fin_url)
# fin_url='https://www.flipkart.com/search?q=apple%20ipad%20pro%202020'
```

```python
# Make a request to the server and read the html content form the url
# if the reponse is 200 then the request is sucessful or else it will show 404 or 500 error
uClient=uReq(fin_url)
flipkartPage=uClient.read()
uClient.close()
flipkart_html=bs(flipkartPage,'html.parser')
```

```python
# the find commentbox form the html using inspect the website
bigboxes=flipkart_html.find_all('div',{'class':'_1AtVbE col-12-12'})
bigboxes
```
